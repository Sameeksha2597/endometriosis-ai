```python
module1_results = [
    ("LogisticRegression", 0.635, 0.657),
    ("RandomForest", 0.598, 0.610),
    ("XGBoost", 0.631, 0.644)
]

for model_name, acc, roc in module1_results:

    with mlflow.start_run(
        run_name=f"Module1_{model_name}"
    ):

        mlflow.log_param("module", "Structured Prediction")
        mlflow.log_param("model", model_name)

        mlflow.log_metric("accuracy", acc)
        mlflow.log_metric("roc_auc", roc)
```
