```python
from PIL import Image

sample_image = os.path.join(
    "../data/external/glenda_enid/endometriosis",
    os.listdir(
        "../data/external/glenda_enid/endometriosis"
    )[0]
)

img = Image.open(sample_image)

print("Size:", img.size)
print("Mode:", img.mode)
```

```python
import random
from PIL import Image
import matplotlib.pyplot as plt

folder = "../data/external/glenda_enid/endometriosis"

samples = random.sample(
    os.listdir(folder),
    6
)

plt.figure(figsize=(12,6))

for i, file in enumerate(samples):
    img = Image.open(
        os.path.join(folder, file)
    )

    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.axis("off")

plt.tight_layout()
plt.tight_layout()

plt.savefig(
    "../docs/figures/glenda_sample_images.png"
)

plt.show()
plt.show()
```
